```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\sairam\Downloads\food nutritional values.csv")
data.head()
```

There are different datatypes in the given dataset like int, float and object. Object datatype refers to the descriptive texts. These are not needed for prediction or other ML-tasks. Hence we can remove all the object datatypes that are Shrt_Desc, GmWt_Desc1 and GmWt_Desc2. Also, index and NDB_No are not necessary for ML tasks. Hence we can remove all these five columns.

```python
data=data.drop(['index'],axis=1)
data=data.drop(['NDB_No'],axis=1)
data=data.drop(['Shrt_Desc'],axis=1)
data=data.drop(['GmWt_Desc1'],axis=1)
data=data.drop(['GmWt_Desc2'],axis=1)
data.info()
```

```python
data.isnull().sum()
```

We can observe that this dataset contains a lot of empty values. Therefore we need to fill them. The nutritional values for these food items would be unique, hence it wont be good to duplicate the values from previous/next data. So using, ffill and bfill are ruled out. The other left out options are using statistical metrics like mean, std, median or interpolation. I am going with the stastical based filling, since interpolation also depends on neighbouring data. Hence I am going to fill all the missing values with the mean of their corresponding column.

```python
data.describe()
```

**Treating empty cells**

```python
a=data.mean()
data=data.fillna(a)
```

```python
data.isnull().sum()
```

All the empty values have been filled with the mean. The next step is to check for any duplicates and removing them.

**Dealing with duplicates**

```python
data.duplicated()
data.drop_duplicates()
```

There were 27 duplicate values which have been removed. The next step is to detect outliers in the dataset. For this I am going to use the Z-Score metric. Z-Score calculates the distance of a given datapoint to the mean with respect to standard deviation. It is proved that 99.7% of data falls into the distribution where the S.D is less than three. So any values beyond them are considered as outliers.

**Dealing with outliers**

```python
from scipy import stats
for x in data.columns:
    z = np.abs(stats.zscore(data[x]))
    out=np.where(z>3)[0]
    print(x+ " "+str(out))
    data[x]=data[x].drop(out)
```

```python
data=data.dropna()
```

Hence the empty data with which we filled using the mean, acts as outliers. But are they really outliers? The answer to this question is unknown and it requires domain knowledge. What I would suggest is that we can train ML algorithm with two datasets, the one which has the entire data, and another in which the outliers are removed. Based on the performance, we can determine whether the removal of outliers is needed or not. Now we have 5032 rows with data that dont have outliers.

The last step in data cleaning is feature scaling. This comprises of two methods, namely, standardisation and normalization. The use of either standardisation or nomalization depends on the type of ML algorithm to be used. In this case, i will perform both standardisation and normalization.

**Normalization**

```python
data_norm=data.copy()
for column in data.columns:
    data_norm[column] = (data_norm[column] - data_norm[column].min()) / (data_norm[column].max() - data_norm[column].min()) 
data_norm.head()
```

**Standardisation**

```python
data.describe()
```

```python
from sklearn.preprocessing import StandardScaler

# copy of datasets
data_stand = data.copy()
# apply standardization on numerical features
for i in data_stand.columns:
    
    # fit on training data column
    scale = StandardScaler().fit(data_stand[[i]])
    
    # transform the training data column
    data_stand[i] = scale.transform(data_stand[[i]])
data_stand.head()
```
